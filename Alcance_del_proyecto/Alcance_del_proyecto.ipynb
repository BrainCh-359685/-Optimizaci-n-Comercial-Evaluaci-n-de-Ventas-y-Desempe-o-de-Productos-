# Alcance del trabajo

### Descripción del Problema
El proyecto tiene como objetivo realizar un análisis exhaustivo de los datos de ventas para proporcionar insights estratégicos que ayuden a mejorar la toma de decisiones comerciales. A través de la evaluación de distintos aspectos del comportamiento de las ventas, se busca identificar patrones, oportunidades y áreas de mejora dentro de la organización.

### Pregunta clave a responder
1. ¿Cuáles son los 5 productos más vendidos en términos de unidades y valor?

2. ¿Existen productos con bajo rendimiento? 

3. ¿Que paises tienen mayores ventas?

### Metodología para el análisis
1. Limpieza y preprocesamiento de datos
2. Análisis exploratorio
4. Conclusiones

### Resultados

#### 1. Limpieza de datos

- Se eliminaron datos faltantes
- Se eliminaron registross de filas repetidas
- Se eliminaron valores extremos ("outliers")

El conjunto de datos original contenía 2823 filas y 25 columnas. Tras el proceso de limpieza de datos, el conjunto de datos resultante quedó con 2823,189 filas y 21 columnas.
#### 2. Análisis exploratorio
¿Cuáles son los 5 productos más vendidos en términos de unidades y valor?

- Agrupar los datos por producto.

- Suma de las unidades vendidas y valor de ventas para cada producto

- Ordenar los productos por unidades vendidas y valor de ventas, de mayor a menor.

- Visualización de los 5 productos más vendidos usando gráficos de barras para facilitar la comparación.

 ¿Existen productos con bajo rendimiento?

 - Definir umbral de bajo rendimiento (percentil 25)

 - Agrupar la cantidad de orden por producto.

 - Filtrar los productos que estan por debajo de ese umbral

¿Que paises tienen mayores ventas?
- Agrupar los datos por pais

- Ordenar los paises por total de ventas, de mayor a menor

- Visualización de la venta por pais usando un grafico de barra. Para identificar fácilmente los países con mayores ventas
